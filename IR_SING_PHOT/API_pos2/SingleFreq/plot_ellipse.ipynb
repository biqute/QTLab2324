```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
coeffs = []
names = []
path = r'C:\Users\ricca\Desktop\MAGISTRALE\QTLab2324\IR_SING_PHOT\API_pos2\SingleFreq\files\\'
for i,file in enumerate(os.listdir(path)):
    names.append(file.strip('Mixer1135_.txt'))
    with open(path+file,'r') as f:
        dic = eval(f.read())
        temp = []
        for key, items in zip(dic.keys(),dic.items()):
            temp.append((items[1]))
        coeffs.append(temp)
```

```python
x = np.arange(-0.05,0.05,0.001)
```

```python
x = x.reshape(len(x),1)
y = x
```

```python
X, Y = np.meshgrid(x*0.05, y*0.1)
```

```python
for i in range(len(coeffs)):
    Z = coeffs[i][0] * X**2 + coeffs[i][1] * X * Y + coeffs[i][2] * Y**2 + coeffs[i][3] * X + coeffs[i][4] * Y 

    # Plotting the contour where the equation equals zero (the ellipse)
    plt.contour(X*1/0.05, Y*1/0.1, Z, levels=[0],label=names[i])
plt.xlabel('I [mV]')
plt.ylabel('Q [mV]')
plt.title('Ellipse Plot')
plt.grid(True)
plt.axis('equal')  # To ensure the plot has equal scaling
plt.legend()
```
