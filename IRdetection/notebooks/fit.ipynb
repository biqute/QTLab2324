```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import random
import time
```

```python
vec = []
for i in range(-10,11):
    vec.append(i)
vec = np.array(vec)
y = np.power(vec, 2)
```

```python
plt.plot(vec,y,'.')
plt.show()
```

```python
random.seed(time.time())
y2 = []
for i in range(len(vec)):
    y2.append(random.uniform(-0.1, 0.1)*y[i] + y[i])
```

```python
plt.plot(vec, y, '.', label='Segnale')
plt.plot(vec, y2, '.', label='Rumore+Segnale')
plt.axvspan(-7.5,7.5, alpha=1, color='yellow')
plt.title('Segnale con rumore?')
plt.xlabel('tempo [$\mu$sec]')
plt.ylabel('intensità [cd]')
plt.legend()
plt.grid()
```

```python
def pol2(x,a,b,c):
    return a * np.power(x,2) + b*x + c
```

```python
popt, _  = curve_fit(pol2,vec,y2)
```

```python
a, b, c = popt
plt.plot(vec, pol2(vec, a, b, c), label = 'pol2')
plt.plot(vec, y2, 'o', label='Rumore+Segnale')
plt.axvspan(-7.5,7.5, alpha=1, color='yellow')
plt.title('Segnale con rumore?')
plt.xlabel('tempo [$\mu$sec]')
plt.ylabel('intensità [cd]')
plt.legend()
plt.grid()
```

```python
fig, ax = plt.subplots(1,2)
ax[0].plot(vec, y2, 'x', label='Signal')
ax[0].plot(vec, pol2(vec, a, b, c), label = 'pol2')
ax[1].plot(vec, y-y2, '.', label='Noise')
ax[0].legend()
ax[1].legend()
```
