```python
def resonance_model(x, Q, Qc,phi0):
    return 1-Q/Qc*np.exp(1j*phi0)/(1+1j*((2*x*Q)))

def resonance_model2(x, Q, Qc,phi0):
    return abs(1-Q/Qc*np.exp(1j*phi0)/(1+1j*((2*x*Q))))

def model(x, m, q, Q, Qc,phi0):
    params = np.array([m,q])
    pol = np.polyval(params, x)
    res = resonance_model(x, Q,Qc,phi0)
    return np.abs(res)*pol

def FWHM(x,y):
    x = np.array(x)
    y = np.array(y)
    half = np.max(y) - (np.max(y)-np.abs(y)) / 2.
    minimums = x[np.where(y < half)]
    return (max(minimums) - min(minimums))

def mod(x,y):
    return np.sqrt(x**2 + y**2)
    
def Q_stima(x0,w):
    if (((x0 is None) or (w is None)) or w==0):
        print('Q_stima: None detected!')
        return 0
    else:
        return x0/(2*w)
    
def Sort_Tuple(tup):
 
    # getting length of list of tuples
    lst = len(tup)
    for i in range(0, lst):
        for j in range(0, lst-i-1):
            if (tup[j][1] > tup[j + 1][1]):
                temp = tup[j]
                tup[j] = tup[j + 1]
                tup[j + 1] = temp
    return tup

def read_files(dir = r'F:\\LabIV\\QTLab2324\\IRdetection\\Instruments\\Test_data\\data_20_02_set1\\'):

    dfs, dictionaries, temps = [], [], []

    for folder in os.listdir(dir):
        if ('T' in folder):
            temps.append(int(str(folder).replace('T_','')))
            parent = dir+folder+'\\'
            dictionaries = []
            for (i,filename) in enumerate(os.listdir(parent)):
                my_dest = filename.replace('.h5','.hdf5')
                my_source = parent + filename
                my_dest = parent + my_dest
                os.rename(my_source, my_dest)
                with h5.File(my_dest, 'r') as r:
                    group = r.require_group(name='raw_data')
                    dic = dict()
                    for c in group.items():
                        dic[c[0]] = c[1][:]
                dictionaries.append(dic)
        dfs.append(dictionaries)
    return dfs, temps
```

Data are a bit noisy $\rightarrow$ Moving Average Filter

```python
def moving_average(data, window_size):
    """
    Apply a moving average smoothing to the input data.
    
    Parameters:
        data (ndarray): Input data array (e.g., S21 versus frequency).
        window_size (int): Size of the moving average window.
        
    Returns:
        ndarray: Smoothed data.
    """
    # Pad the data to handle edge effects
    padded_data = np.pad(data, (window_size // 2, window_size // 2), mode='edge')
    
    # Apply the moving average filter
    smoothed_data = np.convolve(padded_data, np.ones(window_size) / window_size, mode='valid')
    
    return smoothed_data

# Generate example S21 versus frequency data (replace with your actual data)
frequency = dfs[0][0]['f']  # Frequency vector
s21 = dfs[0][0]['S21']

# Apply moving average smoothing
window_size = 10
smoothed_s21 = moving_average(s21, window_size)
# Adjust the frequency range to match the length of the smoothed data
frequency_smoothed = frequency[window_size//2:-window_size//2]
smoothed_s21 = smoothed_s21[:-1]
frequency = (frequency-frequency[np.argmin(smoothed_s21)])/frequency[np.argmin(smoothed_s21)]
plt.scatter(frequency,s21,s=1, label='Raw Data', color='black')
plt.scatter(frequency,smoothed_s21,s=1, label='Smoothed Data', color='red')
plt.xlabel(r'$\nu$ [Hz]')
plt.ylabel(r'$|S_{21}|$ [dB]')
plt.legend()
fig=plt.gca()
fig.set_facecolor('bisque')
plt.grid()
```

```python
m, q = np.polyfit(frequency, smoothed_s21,1)
par0 = [m, q, 1.3e3, 1.2e3,-0.02] 
rms0 = np.sqrt(np.mean(np.square(smoothed_s21[np.where(frequency<-0.001)])))

plt.scatter(frequency,model(frequency,*par0),s=1, label='Expected', color='red')
plt.scatter(frequency,smoothed_s21,s=1, label='Filtered data', color='black')
plt.xlabel(r'$\frac{\nu-\nu_{0}}{\nu_{0}}$ ')
plt.ylabel(r'$|S_{21}|$ [dB]')
plt.legend()
fig=plt.gca()
fig.set_facecolor('bisque')
plt.grid()
```

```python
lsq0 = cost.LeastSquares(frequency, smoothed_s21, rms0, model)
m0 = Minuit(lsq0, *par0)
m0.migrad();

plt.scatter(frequency,smoothed_s21,s=1,marker='.', label='Smoothed Data', color='black')
plt.scatter(frequency,model(frequency,*m0.values[:]),s=1, label='Fit', color='red')
#plt.fill_between(f, f-rms, f+rms, alpha=0.5, color='blue')
plt.xlabel(r'$\frac{\nu- \nu_{0}}{\nu_{0}}$')
plt.ylabel(r'$|S_{21}|$ [dB]')
fig = plt.gca()
fig.set_facecolor('bisque')
plt.grid()

chi2 = m0.fval
ndof = m0.ndof
title = [
            f"$\chi^{2}_{0}$ = {chi2/ndof:.4f}",
        ]

for par in m0.parameters:
    title.append(
        f"{par} = {m0.values[par]:.4f} +/- {m0.errors[par]:.4f}"
    )
plt.legend(title="\n".join(title))
plt.title('Resonance n0')
```

We have to estimate $\sigma$\
To do this we will grab the first 100 data points and get the _RMS_

```python
rms = []
for i in range(len(dfs)):
    for j in range(4):
        condition = lambda x: (x<min_freq[i*4+j]*(1-0.9)*1e9)
        fs = [x for x in dfs[i][j]['f'] if condition(x)]
        filtered_indices = [k for k, x in enumerate(dfs[i][j]['f']) if condition(x)]
        S21s = [dfs[i][j]['S21'][k] for k in filtered_indices]

        order = 1 
        pol_amp = np.polyfit(fs, S21s, order)
        rms.append(np.sqrt(np.mean((np.polyval(pol_amp, fs)-S21s)**2)))
```

**Let's use data around the resonance minimum**

```python
for i in range(len(dfs)):
    for j in range(4):
        condition2 = lambda x: (x>min_freq[i*4+j]*(1-0.02) and x<min_freq[i*4+j]*(1+0.02))
        filtered_indices2 = [i for i, x in enumerate(dfs[i][j]['f']) if condition2(x)]
        f = [x for x in dfs[i][j]['f'] if condition2(x)]
        S21 = [dfs[i][j]['S21'][k] for k in filtered_indices2]
```

```python
for i in range(len(dfs)):
    for j in range(4):
        dfs[i][j]['f'] = (dfs[i][j]['f']-min_freq[i*4+j])/min_freq[i*4+j]
        dfs[i][j]['S21'] = dfs[i][j]['S21']/max(dfs[i][j]['S21'])
```

Now we have to estimate resonance parameters: $Q$, $Q_{c}$, $\phi_{0}$

```python
p0 = []
for i in range(len(dfs)):
    for j in range(4):
        m, q = np.polyfit(dfs[i][j]['f'], dfs[i][j]['S21'],1)
        p0.append([m, q, 1.3e4, 1e3,-0.077])
```

```python
lsq, m = [], []
for i in range(len(dfs)):
    for j in range(4):
        lsq.append(cost.LeastSquares(dfs[i][j]['f'], dfs[i][j]['S21'], rms[i*4+j], model))
        m.append(Minuit(lsq[i*4+j], *p0[i*4+j]))
        m[i*4+j].migrad(ncall=10000)
```
