```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.array([0.12566, 0.12500, 0.12566, 0.12937, 0.13064, 0.13902, 0.13881, 0.13712])
y = np.array(range(len(x)))
s = np.array([0.00040, 0.00135, 0.00075, 0.00282, 0.00282, 0.00121, 0.00177, 0.00147])
```

```python
d1 = sum(np.array(x[1::3])/np.array(s[1::3])**2)/sum(1/np.array(s[1::3])**2)
s1 = np.sqrt(1/sum(1/np.array(s[1::3])**2))
d2 = sum(np.array(x[4::7])/np.array(s[4::7])**2)/sum(1/np.array(s[4::7])**2)
s2 = np.sqrt(1/sum(1/np.array(s[4::7])**2))
d =  np.array([d1, d2])
s =  np.array([s1, s2])
```

```python
plt.figure(figsize=(10,5))
plt.scatter(d,[-1,1], marker='x', color='red', label=r'$\Delta_{0}$  extracted')
plt.errorbar(d,[-1,1],s,np.zeros(len(d)),linestyle='')
plt.legend()
```
